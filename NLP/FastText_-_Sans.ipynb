```python
#The Mahabharata corpus is taken form the following website:
#https://bombay.indology.info/mahabharata/welcome.html
data=[]
n=19 #18 Texts of Mahabharatha
for i in range(1,n):
    path="SanskritMB/MBh"+str(i) + '.txt'
    #path="mb/maha"+str(i) + '.txt'
    #print(path)
    file = open(path)
    lines = file.readlines()
    i=0
    pre=[]
    next=[]
    for sent in lines:
      temp=sent.split()
      #print(pre)
      if len(temp)<=3 :
        #print(temp)
        continue
      if i==0:
        pre=sent.split()
        i=i+1
        continue
      else:
        next=sent.split()
      if i==1:
        l=pre[1:]+next[1:]
        data.append(l)
        #print(l)
        i=0
```

```python
model = FastText(data,vector_size=100,window=5)
```

```python
words = ['नारायणं','सरस्वतीं','संजय','युद्ध','गज','देव','k']
for i in range(len(words)):
    print(words[i], end="\t==>  ")
    similar = model.wv.most_similar(words[i], topn = 5)
    for j in range(len(similar)):
        print(similar[j][0],end =", ")
    print("\n")
```
