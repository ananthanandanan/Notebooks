```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
import re
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
files = os.listdir('./data')
files.sort()
print(files)
```

```python
doc = []

for file in files:
    with open('./data/{}'.format(file)) as fd:
        # each line in the file is read
        lines = fd.readlines()
        # each line in the file is append to doc_lines
        doc_lines = ""
        for line in lines:
            doc_lines+=line
        doc.append(doc_lines)
```

```python
docs = np.array(doc)
print(docs)
```

```python
docs_df = pd.DataFrame({"texts":docs})
docs_df["texts"]
```

```python
import string
import re

def preProcessing(text):
    stemm_snow = SnowballStemmer("english")
    lem = WordNetLemmatizer()
    text = text.replace("\n", " ")
    text = re.sub(r"[0-9]", "", text)
    text = text.lower()
    text = word_tokenize(text)
    text = [stemm_snow.stem(word) for word in text]
    text = [lem.lemmatize(word) for word in text]
    text = [char for char in text if char not in string.punctuation]
    text = [word for word in text if word not in stopwords.words('english')]
    text = ' '.join(text)
    
    return text
```

```python
docs_df['texts'] = docs_df['texts'].apply(preProcessing)
```

```python
sentence = []
texts = docs_df['texts']
for i in range(len(texts)):
    sentence.append(texts[i].strip())
```

```python
count_vector = CountVectorizer(sentence)
sparse_matrix = count_vector.fit_transform(sentence)
tokens = count_vector.get_feature_names()
```
