## vocabulary similarity engine

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk import word_tokenize
from nltk.stem.snowball import SnowballStemmer
from nltk.corpus import stopwords
import re
```

```python
text1 = open("./data/001.txt").read()
token_1 = word_tokenize(text1)
print(token_1)
```

```python
## setting stopword
stop_words = set(stopwords.words('english'))
```

```python
## filtering stopwords
filter_token_1 = [w for w in token_1 if not w.lower() in stop_words]
print(len(filter_token_1), len(token_1))
```

```python
## remove punctuations
pattern = '\W|\d|\s'#a digit or non word
res_token = []
remove_space = []
for w in filter_token_1:
    res_token.append(re.sub(pattern,' ', w))
## remove space
remove_space = [x.replace(' ', '') for x in res_token]
## remove empty string in list
word_tokens = [x for x in remove_space if x]
print(word_tokens)
```

```python
##stem token
stemmer2 = SnowballStemmer("english")
for w in word_tokens:
    print(w, '-->',stemmer2.stem(w))
```
